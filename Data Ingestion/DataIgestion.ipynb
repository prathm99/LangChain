```python
## Web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loaderweb = WebBaseLoader(web_paths=('https://jalammar.github.io/illustrated-transformer/',),
                          bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                              class_=('decoder', 'encoder')
                          ))
                          )
loaderweb.load()
```
