```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["LANGCHAIN_API_KEY"] = os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ['LANGCHAIN_PROJECT'] = os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-4o')
print(llm)
```

```python
res = llm.invoke("What is the capital of France?")  # Example usage
print(res)  # Output the result of the invocation
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a expert AI assistant."),
    ("human", "{question}"),
    ("ai", "{answer}"),
])

prompt
```

```python
chain=prompt | llm

chain.invoke({"question": "What is the capital of France?"})  # Example usage
```

```python
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt | llm | output_parser

# without a output parser, the chain will return a LangChain object without proper output, with the help of parser
# we will get clear output
chain.invoke({"question": "What is the capital of France?"})  # Example usage
```
