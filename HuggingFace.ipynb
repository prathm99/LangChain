```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HF_TOKEN'] = os.getenv('HF_TOKEN')
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeds = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
text = "The quick brown fox jumps over the lazy dog."
embeddings = embeds.embed_query(text)
```

```python
embeddings
```

```python
len(embeddings)  # This will give you the dimension of the embeddings
```

```python
doc_res = embeds.embed_documents([text, "This is another document."])
```
