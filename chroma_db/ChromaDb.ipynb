```python
from langchain_chroma import Chroma
```

```python
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import OllamaEmbeddings
```

```python
loader = TextLoader("text.txt")
data = loader.load()
data
```

```python
textsplit = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splits = textsplit.split_documents(data)
```

```python
embeds = OllamaEmbeddings(model="gemma:2b")
vecdb = Chroma.from_documents(documents=splits, embedding=embeds)
vecdb
```

```python
query = "how all this mistaken idea of denouncing pleasure and praising pain was born?"
op = vecdb.similarity_search(query)
```

```python
op[0].page_content
```

```python
vecdb = Chroma.from_documents(documents=splits, embedding=embeds, persist_directory="./chroma_db")
```

```python
retreiver = vecdb.as_retriever()
retreiver.invoke(query)[0]
```
